```python
import numpy as np
import matplotlib.pyplot as plt
```

### 데이터 로드

```python
fruits = np.load('fruits_300.npy')
```

```python
print(fruits.shape)

# 총 300장의 사진이 있는데 사진 1장당 100 x 100 픽셀
```

```python
# 흑백사진은 0부터 255 까지 정수값을 가짐



print(fruits[0, 0, :])
```

```python
plt.imshow(fruits[0], cmap='gray')
plt.show()

# 위에껀 첫번째 줄 하나로 이루어진것  중간에 숫자 커지는 부분 윗꼭지 부분

#흑백사진은 0에 가까울 수록 검게 나타남
#사과사진
```

```python
plt.imshow(fruits[0], cmap='gray_r')
plt.show()


#원본 사진은 이게 정상 -> 사과에 집중해야하는데 바탕색에 집중하게됨 -> 바탕색을 검게 된 걸 공부하는 이유
```

```python
fig, axs = plt.subplots(1, 2)
axs[0].imshow(fruits[100], cmap='gray_r')
axs[1].imshow(fruits[200], cmap='gray_r')

# 파인애플 과 바나나
```
